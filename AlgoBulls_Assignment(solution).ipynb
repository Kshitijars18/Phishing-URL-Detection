**Task 3:** Define a class Strategy, which can do the following, given a script name

```python
import pandas as pd
import requests

class ScriptData:
    def __init__(self):
        self.intraday_data = {}

    def fetch_intraday_data(self, script):
        API_KEY = 'SPEQGVGL40TIZHCM'
        function = 'TIME_SERIES_INTRADAY'
        interval = '1min'
        symbol = script
        url = f'https://www.alphavantage.co/query?function={function}&symbol={symbol}&interval={interval}&apikey={API_KEY}'
        response = requests.get(url)
        self.intraday_data[script] = response.json()

    def convert_intraday_data(self, script):
        data = self.intraday_data[script]['Time Series (1min)']
        timestamp = [pd.Timestamp(time) for time in data.keys()]
        open_price = [float(data[time]['1. open']) for time in data.keys()]
        high = [float(data[time]['2. high']) for time in data.keys()]
        low = [float(data[time]['3. low']) for time in data.keys()]
        close = [float(data[time]['4. close']) for time in data.keys()]
        volume = [int(data[time]['5. volume']) for time in data.keys()]
        self.intraday_data[script] = pd.DataFrame({'timestamp': timestamp, 'open': open_price, 'high': high, 'low': low, 'close': close, 'volume': volume})

    def __getitem__(self, script):
        return self.intraday_data[script]

    def __setitem__(self, script, df):
        self.intraday_data[script] = df

    def __contains__(self, script):
        return script in self.intraday_data

    def indicator1(df, timeperiod):
       df['indicator'] = df['close'].rolling(timeperiod).mean()
       return df[['timestamp', 'indicator']]

class Strategy:
    def __init__(self, script_name):
        self.script_name = script_name
        self.script_data = ScriptData()
        
    def get_script_data(self):
        self.script_data.fetch_intraday_data(self.script_name)
        self.script_data.convert_intraday_data(self.script_name)
        self.df = self.script_data[self.script_name]
        
    def get_signals(self, timeperiod=5):
        self.indicator_data = indicator1(self.df, timeperiod)
        close_data = self.df['close']
        signals = []
        for i in range(1, len(self.df)):
            if self.indicator_data['indicator'][i] > close_data[i] and self.indicator_data['indicator'][i-1] <= close_data[i-1]:
                signals.append({'timestamp': self.df['timestamp'][i], 'signal': 'BUY'})
            elif self.indicator_data['indicator'][i] < close_data[i] and self.indicator_data['indicator'][i-1] >= close_data[i-1]:
                signals.append({'timestamp': self.df['timestamp'][i], 'signal': 'SELL'})
        self.signals = pd.DataFrame(signals)
        print(self.signals[self.signals['signal'].isin(['BUY', 'SELL'])])
    
```

```python
strategy = Strategy('NVDA')
strategy.get_script_data()
strategy.get_signals()
```
